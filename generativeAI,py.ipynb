```python
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
test(1,2,3,4,5,6)
```

```python
test(4,8,9,7,11,8,54,459,)
```

```python
def sum_number(*args):
  sum=0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(789,952,741,963,759,)
```

```python
# keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a=10,b=20,c=30)
```

```python
#cosine similarirty
import numpy as np
from numpy.linalg import norm
a= np.array([2,1,2,3,2,9])
b= np.array([3,4,2,4,5,5])

cosine = np.dot(a,b)/(norm(a)*norm(b))
print("cosine similarity:",cosine)
```

```python
from transformers import pipeline
summarizer = pipeline("summarization",model="facebook/bart-large-cnn")

text = """the global shift toward renewable energy is accelerating, driver by concerns over climate change and energy security. solat and wind power installations have reached record levels worldwide."""
summary = summarizer(text,max_length=30,min_length=10,do_sample=False)
print(summary)
```
